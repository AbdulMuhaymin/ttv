```python
from airium import Airium
from pandas import read_csv, read_fwf
from glob import glob

planets = glob("data/mod_data/*.txt")

for planet in planets:
    
    end = planet.find('_oc')
    if planet[14:18] == "hatp":
        name = "HAT-P-" + planet[18:end]
    elif planet[14:18] == "wasp":
        name = "WASP-" + planet[18:end]
    
    root_df = read_fwf(planet, skiprows=1, header=None, sep="\t+", dtype="str")
    root_df = root_df.drop(root_df[root_df[8] == 'ThisStudy'].index)
    
    all_dt_pts = glob("data/exo/" + name + "/**/**/")
    for i in all_dt_pts:
        if "Database" in i:
            ind = 27
        elif "Literature" in i:
            ind = 29
        
        loc = i + "/exofast/exofast.usr.param.txt"
        temp = read_fwf(loc, skiprows=7, header=None, sep="\t+", dtype="str")
        jd = temp.iloc[19].iat[6]

        for k in range(len(root_df)):
            if root_df.iat[k, 0] == jd:
                MidTrTime = root_df.iat[k, 0]
                MidTrTimeErr = root_df.iat[k, 1]
                link = root_df.iat[k, 23]

                
                a = Airium()
                a('<!DOCTYPE html>')
                with a.html(lang="en"):

                    with a.head():
                        a.meta(charset="utf-8")
                        a.link(rel="stylesheet", href="https://cdn.jsdelivr.net/npm/bootstrap@4.4.1/dist/css/bootstrap.min.css")
                        with a.title():
                            a("Homogeneous Transit Timing Analyses of Ten Exoplanet Systems")
                        with a.style():
                            a(".responsive {width: 100%; height: auto; max-width: 891px;}")

                    with a.body():

                        with a.h2():
                            a("EXOFAST Model for " + name)

                        with a.div():
                            with a.p():
                                with a.h4():
                                    a("Measured Mid-Transit Time in BJD-TDB: ")
                                    a(MidTrTime)
                                a.br()
                                with a.h4():
                                    a("Error of the Measured Mid-Transit Time: ")
                                    a(MidTrTimeErr)
                            with a.button(type="button", klass="btn btn-success", onclick="window.open('" + link + "', '_blank')"  ):
                                a("Click here to view the data source")

                        a.img(src= 'preview.png', klass="responsive", alt= "Sorry, the image couldn't load properly. If you see this message, please let the administrator know.")

                #html = str(a)
                html = bytes(a)
                with open(i + 'preview.html', 'wb') as f:
                    f.write(bytes(html))
```
